```python
data = TokenDataset('./input_text_cleaned.txt', line_by_line=True)
```

```python
from aitextgen import aitextgen
```

```python
# ai = aitextgen(tf_gpt2="124M",  to_gpu=True)
ai = aitextgen(model=model,  to_gpu=True)
```

```python
ai.train('input_text_cleaned.txt',
         line_by_line=True,
         from_cache=False,
         num_steps=500,
         generate_every=100,
         save_every=500,
         save_gdrive=False,
         learning_rate=1e-3,
         fp16=False,
         batch_size=1, 
         )
```

```python
ai.save()
```
