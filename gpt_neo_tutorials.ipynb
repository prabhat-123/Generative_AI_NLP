### Import Pipeline from transformers

```python
from transformers import pipeline
```

### Setup generator

```python
generator = pipeline('text-generation', model='EleutherAI/gpt-neo-1.3B')
```

### Generate Text Using Prompts

```python
prompt = "What is the meaining of life?"
```

<li>generator takes in four parameters.</li>
<li>It takes in input the prompt, the maximum length of output it is going to generate.</li>
<li>do_sample=True, it allow us to leverage sampling within our model</li>
<li>temperature; this influences how it actually determines results going forwards from a particular token that it actually had returned back.</li>

```python
result = generator(prompt, max_length=50, do_sample=True, 
                   temperature=0.9)
```

### Output Text

```python
result[0]['generated_text']
```
